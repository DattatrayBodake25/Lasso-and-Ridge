# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso,Ridge
```

```python
x = df.iloc[:,:-1]
y = df['price']
```

```python
x.head()
```

```python
y.head()
```

```python
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
#splitting data
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size=0.25,random_state=1)
```

# Model Building

# 1) Linear Regression

```python
lr = LinearRegression()
lr.fit(xtrain,ytrain)
ypred = lr.predict(xtest)
print('Training Accuracy: ',lr.score(xtrain,ytrain))
print('Testing Accuracy: ',lr.score(xtest,ytest))
```

```python
lr.coef_
```

 - Training accuracy is high - low bias
 - Testing accuarcy is low - high variance
 
**Overfitting**

# 1) Ridge Regression

```python
rg = Ridge()
rg.fit(xtrain,ytrain)
rgpred =rg.predict(xtest)
print('Training Accuracy: ',rg.score(xtrain,ytrain))
print('Testing Accuracy: ',rg.score(xtest,ytest))
```

```python
rg.coef_
```

# 2) Lasso Regression

```python
ls = Lasso()
ls.fit(xtrain,ytrain)
ls_pred = ls.predict(xtest)
print('Training Accuracy: ',ls.score(xtrain,ytrain))
print('Testing Accuracy: ',ls.score(xtest,ytest))
```

```python
ls.coef_
```

# Selecting of Alpha

```python
for i in range(50,200):
    rg = Ridge(alpha=i)
    rg.fit(xtrain,ytrain)
    y = rg.predict(xtest)
    
    print(f"{i} {rg.score(xtrain,ytrain)} {rg.score(xtest,ytest)}")
```

```python
for i in range(3000,4000):
    ls = Lasso(alpha=i)
    ls.fit(xtrain,ytrain)
    y = ls.predict(xtest)
    
    print(f"{i} {ls.score(xtrain,ytrain)} {ls.score(xtest,ytest)}")
```
